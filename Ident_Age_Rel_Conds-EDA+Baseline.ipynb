## Modeling 

### First we try XGBoost ...

We tune our XGBoost model using a manual Grid-Search with 5-Fold C.V. This will make the most out of the limited data, providing a balanced log-loss estimate that has a smaller variance. Manully seemed easier due to the fact that I want to over-sample AND shuffle before training AND use a custom loss function.

```python
def bal_log_loss(p, y):
    ind0 = np.where(y==0)[0]
    ind1 = np.where(y==1)[0]
    
    N0 = len(ind0)
    N1 = len(ind1)
    
    y0 = (y==0).astype(int)
    y1 = y.astype(int)
    
    return (- np.sum(y0*np.log(p[:, 0]))/N0 - np.sum(y1*np.log(p[:, 1]))/N1) / 2
```

```python
# setup dataframes for predictors and response
X = train_df.drop(columns=['Class', 'AgeRelCond', 'Id'])
X = pd.get_dummies(X, drop_first=True)

y = train_df['Class'].astype(int)

# set verbose
verbose = True

# set fold size (5 fold cv)
fold_sz = np.floor(X.shape[0] / 5).astype(int)

# inti total performance metrics lists
accs_total = []
blls_total = []
blls_train_total = []

# loop through each fold
for i in range(5):
    
    print(f"\n*********************** FOLD #{i} ***********************\n")
    
    mask = np.ones(X.shape[0]).astype(bool)
    mask[(i*fold_sz):((i+1)*fold_sz)] = 0
    
    X_train_raw = X.to_numpy()[mask,]
    y_train_raw = y.to_numpy()[mask,]
    
    X_val = X.to_numpy()[~mask,]
    y_val = y.to_numpy()[~mask,]
    
    # over sample the diagnosed patients in training set
    ros = RandomOverSampler(random_state=77, sampling_strategy='minority')
    X_train, y_train = ros.fit_resample(X_train_raw, y_train_raw)

    # shuffle (in case the model choice may be impacted by ordering)
    shuff_ind = np.random.choice(len(y_train), len(y_train), replace=False)

    X_train = X_train[shuff_ind,]
    y_train = y_train[shuff_ind,]
    
    # show old vs new balance
    print(f'\nResampling ...')
    print(f'Raw training data shape = {Counter(y_train_raw)}')
    print(f'Resampled training data shape = {Counter(y_train)}\n\n')
    

    # set up grid of hyper-params
    parameters = {
        'max_depth': range(2, 8, 2),
        'n_estimators': range(100, 300, 100),
        'learning_rate': np.linspace(0.5, 1, num=25)
    }

    # init lists of metrics
    accs = []
    blls = []
    blls_train = []

    # init counter
    trial = 1

    for n in parameters['max_depth']:
        for m in parameters['n_estimators']:
            for k in parameters['learning_rate']:

                xgb_classifier = xgb.XGBClassifier( max_depth=n, n_estimators=m, eta = k)
                xgb_classifier.fit(X_train, y_train)
                
                # collect train probabilities 
                probs_tr = xgb_classifier.predict_proba(X_train)
                
                # collect train balanced logarithmic loss
                bll_train = bal_log_loss(probs_tr, y_train)
                
                # store train metric
                blls_train.append(bll_train)
                

                # collect val probabilities 
                probs = xgb_classifier.predict_proba(X_val)

                # collect val accuracy 
                acc = np.mean(np.argmax(probs, 1) == y_val)
                
                # collect val balanced logarithmic loss
                bll = bal_log_loss(probs, y_val)

                # store val metrics
                accs.append(acc)
                blls.append(bll)

                if verbose:
                    print(f'Trail #{trial:2} || Using: max_depth={n}, n_estimators={m}, eta={k:.4f} || Balanced Log-Loss = {bll:.5f}')

                trial += 1
            
    accs_total.append(accs)
    blls_total.append(blls)
    blls_train_total.append(blls_train)
    
```

```python
# store each fold's metrics in array
accs_total = np.array(accs_total)
blls_total = np.array(blls_total)


accs_CV_est = np.mean(accs_total, axis=0)
blls_CV_est = np.mean(blls_total, axis=0)
blls_train_CV_est = np.mean(blls_train_total, axis=0)

print(f'Trial #{np.argsort(blls_CV_est)[0]+1} had the lowest balanced log-lossat at {np.sort(blls_CV_est)[0]}.')


figure, axis = plt.subplots(3, 1, figsize=(10, 10))
  
axis[0].plot(accs_CV_est)
axis[0].set_title("5 Fold-CV Validation Accuracy Est.")
  
axis[1].plot(blls_CV_est)
axis[1].set_title("5 Fold-CV Validation Balanced Log-Loss Est.")

axis[2].plot(blls_train_CV_est)
axis[2].set_title("Average Training Balanced Log-Loss Est.")

plt.show()
```

Trial \#76 seems to have the lowest (and simplest) balanced log-loss across our validation sets. This corresponds to a max_depth=4, n_estimators=200, and eta=0.5000. The final estimated balanced log-loss was 0.37915.

### Using PCA for dimension reduction before applying XGBoost

Note that we must scale the data before using PCA to allow for an equal consideration of each feature.

```python
# setup dataframes for predictors and response
X = train_df.drop(columns=['Class', 'AgeRelCond', 'Id'])
X = pd.get_dummies(X, drop_first=True)

y = train_df['Class'].astype(int)

# set verbose
verbose = True

# set fold size (5 fold cv)
fold_sz = np.floor(X.shape[0] / 5).astype(int)

# inti total performance metrics lists
accs_total = []
blls_total = []
blls_train_total = []

# loop through each fold
for i in range(5):
    
    print(f"\n*********************** FOLD #{i} ***********************\n")
    
    mask = np.ones(X.shape[0]).astype(bool)
    mask[(i*fold_sz):((i+1)*fold_sz)] = 0
    
    X_train_raw = pd.DataFrame(X.to_numpy()[mask,], columns=X.columns)
    y_train_raw = pd.Series(y.to_numpy()[mask,])
    
    X_val = pd.DataFrame(X.to_numpy()[~mask,], columns=X.columns)
    y_val = pd.Series(y.to_numpy()[~mask,])
    
    X_train_raw = X_train_raw.fillna(X_train_raw.mean())
    X_val = X_val.fillna(X_train_raw.mean())
    
    # over sample the diagnosed patients in training set
    ros = RandomOverSampler(random_state=77, sampling_strategy='minority')
    X_train, y_train = ros.fit_resample(X_train_raw, y_train_raw)

    # shuffle (in case the model choice may be impacted by ordering)
    shuff_ind = np.random.choice(len(y_train), len(y_train), replace=False)

    X_train = X_train.iloc[shuff_ind,]
    y_train = y_train.iloc[shuff_ind,]
    
    # show old vs new balance
    print(f'\nResampling ...')
    print(f'Raw training data shape = {Counter(y_train_raw)}')
    print(f'Resampled training data shape = {Counter(y_train)}\n\n')
    
    # define scaler for PCA use
    scaler = StandardScaler()

    # created scaled version of training and testing data
    X_trainS = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
    X_valS = pd.DataFrame(scaler.transform(X_val), columns=X.columns)
    
    # define PCA model to check variance explained
    pca = PCA(n_components=30)

    #fit PCA model to data
    pca.fit(X_trainS)

    PC_values = np.arange(pca.n_components_) + 1
    plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
    plt.title('Scree Plot')
    plt.xlabel('Sample Principal Component')
    plt.ylabel('Variance Explained')
    plt.show()
    
    # set up grid of hyper-params
    parameters = {
        'max_depth': range(2, 8, 2),
        'n_estimators': range(50, 350, 100),
        'learning_rate': np.linspace(0, 1, num=25)
    }

    # init lists of metrics
    accs = []
    blls = []
    blls_train = []

    # init counter
    trial = 1    

    for ncomps in range(12, 23):

        # define optimal PCA model to use for predictions
        pca = PCA(n_components=ncomps)

        # fit PCA model to train data, then transform training and validation data
        X_train = pca.fit_transform(X_trainS)
        X_val = pca.transform(X_valS)


        for n in parameters['max_depth']:
            for m in parameters['n_estimators']:
                for k in parameters['learning_rate']:

                    xgb_classifier = xgb.XGBClassifier( max_depth=n, n_estimators=m, eta = k)
                    xgb_classifier.fit(X_train, y_train)

                    # collect train probabilities 
                    probs_tr = xgb_classifier.predict_proba(X_train)

                    # collect train balanced logarithmic loss
                    bll_train = bal_log_loss(probs_tr, y_train)

                    # store train metric
                    blls_train.append(bll_train)


                    # collect val probabilities 
                    probs = xgb_classifier.predict_proba(X_val)

                    # collect val accuracy 
                    acc = np.mean(np.argmax(probs, 1) == y_val)

                    # collect val balanced logarithmic loss
                    bll = bal_log_loss(probs, y_val)

                    # store val metrics
                    accs.append(acc)
                    blls.append(bll)

                    if verbose:
                        print(f'Trail #{trial:2} || Using: n_comp={ncomps}, max_depth={n}, n_estimators={m}, eta={k:.4f} || Balanced Log-Loss = {bll:.5f}')

                    trial += 1
            
    accs_total.append(accs)
    blls_total.append(blls)
    blls_train_total.append(blls_train)
    
```

```python
# store each fold's metrics in array
accs_total = np.array(accs_total)
blls_total = np.array(blls_total)


accs_CV_est = np.mean(accs_total, axis=0)
blls_CV_est = np.mean(blls_total, axis=0)
blls_train_CV_est = np.mean(blls_train_total, axis=0)

print(f'Trial #{np.argsort(blls_CV_est)[0]+1} had the lowest balanced log-loss at {np.sort(blls_CV_est)[0]}.')


figure, axis = plt.subplots(3, 1, figsize=(10, 10))
  
axis[0].plot(accs_CV_est)
axis[0].set_title("5 Fold-CV Validation Accuracy Est.")
  
axis[1].plot(blls_CV_est)
axis[1].set_title("5 Fold-CV Validation Balanced Log-Loss Est.")

axis[2].plot(blls_train_CV_est)
axis[2].set_title("Average Training Balanced Log-Loss Est.")

plt.show()
```

Trial \#456 seems to have the lowest (and simplest) balanced log-loss across our validation sets. This corresponds to a n_comp=14, max_depth=2, n_estimators=50, and eta=0.2083. The final estimated balanced log-loss was 0.43846.

### Output Probabilties 

For this section, we only generate output for the xgboost model because it seems to have a better estimate for the balanced log-loss. It appears that PCA may ultimately lead to a removal of important information.

We first generate probabilities for the *official* validaiton set with tuned hyperparameters.

```python
# load training data
train_df = pd.read_csv(TRAIN_DATA)

# allocate
X = train_df.drop(columns=['Class', 'Id'])
X = pd.get_dummies(X, drop_first=True)

y = train_df['Class'].astype(int)

# train-validation split 
X_train_raw, X_val, y_train_raw, y_val = train_test_split(X, y, test_size=0.20, random_state=42)

# over sample the diagnosed patients in training set
ros = RandomOverSampler(random_state=77, sampling_strategy='minority')
X_train, y_train = ros.fit_resample(X_train_raw, y_train_raw)

# shuffle (in case the model choice may be impacted by ordering)
shuff_ind = np.random.choice(len(y_train), len(y_train), replace=False)

X_train = X_train.iloc[shuff_ind,]
y_train = y_train.iloc[shuff_ind,]

# make uints just in case 
y_train.astype('uint8')
y_val.astype('uint8')

# use optimal hyperparams
xgb_classifier = xgb.XGBClassifier( max_depth=4, n_estimators=300, eta = 0.5)
xgb_classifier.fit(X_train, y_train)

# collect val probabilities 
probs = xgb_classifier.predict_proba(X_val.dropna())

# store the validation-set predictions in csv format, locally.
pd.DataFrame(probs).to_csv("val_pred_probs/xgboost-tuned.csv", header=True, index=False)
```

Then we generate the probabilites for the test set (for submission).

```python
# use optimal hyperparams
xgb_classifier = xgb.XGBClassifier( max_depth=4, n_estimators=300, eta = 0.5)

# over sample the diagnosed patients in training set
ros = RandomOverSampler(random_state=77, sampling_strategy='minority')
X_rs, y_rs = ros.fit_resample(X, y)

# shuffle (in case the model choice may be impacted by ordering)
shuff_ind = np.random.choice(len(y_rs), len(y_rs), replace=False)
X_rs = X_rs.iloc[shuff_ind,]
y_rs = y_rs.iloc[shuff_ind,]

# fit xgboost model
xgb_classifier.fit(X_rs, y_rs)

# load testing data
test_df = pd.read_csv(TEST_DATA)
test_df['EJ_B'] = (test_df['EJ'] == 'B').astype('int')
X_test = test_df.drop(columns=['Id', 'EJ'])

# collect val probabilities 
probs = xgb_classifier.predict_proba(X_test)

# store the test-set predictions in csv format, locally.
pd.DataFrame(probs).to_csv("test_pred_probs/xgboost-tuned.csv", header=True, index=False)
```
